### Investigation of Outliers

```python
import numpy as np
```

```python
Q1 = np.percentile(df["cnt"], 25)
Q3 = np.percentile(df["cnt"], 75)
IQR = Q3 - Q1 
u_w = min(max(df["cnt"]), Q3 + 1.5*IQR) # upper whisker
print(u_w)
```

```python
df2 = df[df["cnt"] > u_w]
df2[["yr",	"mnth",	"hr",	"workingday",	"temp",	"atemp",	"hum",	"windspeed",	"cnt"]].describe()
# 29% outliers only
```

```python
# hourly grouped series (outliers)
hdf = df2.groupby("hr")["cnt"].mean()
hdf.sort_values(ascending=False, inplace= True)

# hourly df
h_df = pd.DataFrame({"Hour" : hdf.index, "Count" : hdf.values} )
h_df
```

### Prereration of Data and Model Building

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df1 = pd.get_dummies(df, columns = ["season", "weathersit"], drop_first=True)
```

```python
X = df1.drop(columns = ["instant", "dteday","casual", "registered","cnt","holiday","weekday"])
Y = df1["cnt"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
regressor = DecisionTreeRegressor(max_depth=5, random_state=42)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2_s = r2_score(y_test, y_pred)
print(f"MSE = {mse}")
print(f"R_2 Score = {r2_s}")
```

### GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV
```

```python
params_grid = {
    "max_depth" : [3,5,10,None],
    "min_samples_split" : [2,10,20],
    "min_samples_leaf" : [1,5,10],
    "max_features" : [None, "sqrt","log2"]
}

regressor = DecisionTreeRegressor(random_state=42)

grid_search = GridSearchCV(estimator= regressor, param_grid=params_grid, cv = 5, scoring= "neg_mean_squared_error", n_jobs = -1)

grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_

print(f"Best Parameters: {grid_search.best_params_}")

y_pred = best_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)


print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

## Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params_grid_r = {
    "max_depth" : [3,5,10,None],
    "min_samples_split" : [2,10,20],
    "min_samples_leaf" : [1,5,10],
    "max_features" : [None, "sqrt","log2"] 
}

# regressor_R = DecisionTreeRegressor(random_state=42)

R_search = RandomizedSearchCV(estimator=regressor, param_distributions = params_grid_r, cv=5, scoring= "neg_mean_squared_error", n_jobs=-1)

R_search.fit(X_train, y_train)

best_model_R = R_search.best_estimator_

print(f"Best Parameter :{R_search.best_params_}")

y_pred_R = best_model_R.predict(X_test)

mse_R = mean_squared_error(y_test, y_pred_R)
r2_R = r2_score(y_test, y_pred_R)

print(f"mean squared error : {mse_R}")
print(f"r2 Score : {r2_R}")
```
